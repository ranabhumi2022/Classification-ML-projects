```python
#Q12 classification
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
from google.colab import files
uploaded= files.upload()
import os
print(os.listdir())
#load dataset
df=pd.read_csv("heart.csv")
print(df.info())

#  feature marking seperate  feature  X and Y
X=df.drop("HeartDisease",axis=1)
Y=df['HeartDisease']
print(df.head())
X = pd.get_dummies(X, drop_first=True)
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)

# Train the model split dataset into training (70%)
X_train, X_test, Y_train, Y_test=train_test_split(X_scaled,Y,test_size=0.2,random_state=42)
model=LogisticRegression(max_iter=2000)
model.fit(X_train,Y_train)
Y_pred=model.predict(X_test)
# coefficient and intercept
print("intercept:",model.intercept_)
print("coefficent:",model.coef_)
accuracy=accuracy_score(Y_test,Y_pred)
print("accuracy:",accuracy)
cm=confusion_matrix(Y_test,Y_pred)
print("confusion matrix :\n",cm)

# New sample input (same format as dataset)
new_data = {
    'General_Health': 'Very Good',
    'Checkup': 'Within 1 year',
    'Exercise': 'Yes',
    'Skin_Cancer': 'No',
    'Other_Cancer': 'No',
    'Depression': 'No',
    'Diabetes': 'No',
    'Arthritis': 'No',
    'Sex': 'Male',
    'Age_Category': '30-34',
    'Height_(cm)': 170,
    'Weight_(kg)': 70,
    'BMI': 24.2,
    'Smoking_History': 'No',
    'Alcohol_Consumption': 2,
    'Fruit_Consumption': 3,
    'Green_Vegetables_Consumption': 4,
    'FriedPotato_Consumption': 1
}

import pandas as pd

new_df = pd.DataFrame([new_data])

# Same preprocessing
new_df = pd.get_dummies(new_df)

# Match training columns
new_df = new_df.reindex(columns=X.columns, fill_value=0)

# Scale
new_df_scaled = scaler.transform(new_df)

# Predict
prediction = model.predict(new_df_scaled)

print("Prediction (0 = No, 1 = Yes):", prediction)

```

```python
#Q13 classification
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from google.colab import files
uploaded=files.upload()
import os
print(os.listdir())

# i. Read dataset
df = pd.read_csv("student_performance.csv")

# ii. Separate features and target
X = df.drop(["grade", "student_id"], axis=1) # Dropping 'student_id' as it's not a feature for prediction
y = df["grade"]

# Feature Scaling (important for KNN)
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Function to run KNN for different splits
def run_knn(test_size):
    print("\n==============================")
    print(f"Train-Test Split: {int((1-test_size)*100)}:{int(test_size*100)}")

    # iii. Split dataset
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

    # iv. KNN model (K = 5)
    model = KNeighborsClassifier(n_neighbors=5)

    # v. Train model
    model.fit(X_train, y_train)

    # Prediction
    y_pred = model.predict(X_test)

    # vi. Evaluation
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Precision:", precision_score(y_test, y_pred, average='weighted'))
    print("Recall:", recall_score(y_test, y_pred, average='weighted'))
    print("F1 Score:", f1_score(y_test, y_pred, average='weighted'))

    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

    return model

# Run for all splits
model_70 = run_knn(0.3)   # 70:30
model_60 = run_knn(0.4)   # 60:40
model_80 = run_knn(0.2)   # 80:20

# vii. Predict for new data
# Example input (replace with actual feature values)
new_data = [[75, 80, 70, 65]] # These should correspond to 'weekly_self_study_hours', 'attendance_percentage', 'class_participation', 'total_score'
new_data = scaler.transform(new_data)

prediction = model_80.predict(new_data)
print("\nPredicted Result for new data:", prediction)
```

```python
#Q14 classification

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from google.colab import files
uploaded=files.upload()
import os
print(os.listdir())

# Load dataset
df = pd.read_csv("loan_data.csv")
print(df.columns)

# Separate features and target
X = df.drop("loan_status", axis=1)
y = df["loan_status"]
df = df.dropna()

# Encode categorical data
le = LabelEncoder()
for col in X.columns:
    if X[col].dtype == 'object':
        X[col] = le.fit_transform(X[col])

y = le.fit_transform(y)

# Scaling
scaler = StandardScaler()
X = scaler.fit_transform(X)

print("Unique target values:", np.unique(y))
print("Sample target values:", y[:10])

# Train-test split (70:30)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# KNN model
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)

# Prediction
y_pred = model.predict(X_test)

# Evaluation
print("\nLoan Dataset Results:")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred,average='weighted'))
print("Recall:", recall_score(y_test, y_pred,average='weighted'))
print("F1 Score:", f1_score(y_test, y_pred,average='weighted'))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# New prediction
new_sample = [X[0]]
print("Prediction:", model.predict([new_sample[0]]))
```
